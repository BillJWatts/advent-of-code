```python
CUBE_ARRAY = {}
AIR_ARRAY = {}
counted_air_coords = []

pocket_id = iter(range(500))

from dataclasses import dataclass
@dataclass
class Coord:
    x: int
    y: int
    z: int
    def __init__(self, x,y,z):
        self.x = int(x)
        self.y = int(y)
        self.z = int(z)
    def __add__(self, other):
        return Coord(self.x + other.x, self.y + other.y, self.z + other.z)
    def __sub__(self, other):
        return Coord(self.x - other.x, self.y - other.y, self.z - other.z)
    def __hash__(self):
        return hash((self.x, self.y, self.z))

UP = Coord(0,1,0)
DOWN = Coord(0,-1,0)
LEFT = Coord(-1,0,0)
RIGHT = Coord(1,0,0)
FORWARD = Coord(0,0,-1)
BACKWARD = Coord(0,0,1)

@dataclass
class Air:
    def __init__(self,coord):
        self.coord = coord
        self.group_id = None
    
    def set_group_id(self, group_id = None):
        if self.group_id:
            return
        if group_id:
            self.group_id = group_id
        else:
            self.group_id = next(pocket_id)

        up = AIR_ARRAY.get(self.coord+UP,None)
        if isinstance(up, Air):
            up.set_group_id(self.group_id)
        down = AIR_ARRAY.get(self.coord+DOWN,None)
        if isinstance(down, Air):
            down.set_group_id(self.group_id)
        left = AIR_ARRAY.get(self.coord+LEFT,None)
        if isinstance(left, Air):
            left.set_group_id(self.group_id)
        right = AIR_ARRAY.get(self.coord+RIGHT,None)
        if isinstance(right, Air):
            right.set_group_id(self.group_id)
        forward = AIR_ARRAY.get(self.coord+FORWARD,None)
        if isinstance(forward, Air):
            forward.set_group_id(self.group_id)
        backward = AIR_ARRAY.get(self.coord+BACKWARD,None)
        if isinstance(backward, Air):
            backward.set_group_id(self.group_id)
    def __repr__(self) -> str:
        return f"Air({self.coord} group_id: {self.group_id}"
    

class Cube:
    

    def __init__(self, coord):
        self.coord = coord
        self.open_face_coords = []
    
    @property
    def open_faces(self):
        open_faces = 0
        if not CUBE_ARRAY.get(self.coord+UP,None):
            AIR_ARRAY[self.coord+UP] = Air(self.coord+UP)
            if not CUBE_ARRAY.get(self.coord+UP+UP,None):
                AIR_ARRAY[self.coord+UP+UP] = Air(self.coord+UP+UP)
            open_faces += 1
            self.open_face_coords.append(self.coord+UP)
        if not CUBE_ARRAY.get(self.coord+DOWN,None):
            AIR_ARRAY[self.coord+DOWN] = Air(self.coord+DOWN)
            if not CUBE_ARRAY.get(self.coord+DOWN+DOWN,None):
                AIR_ARRAY[self.coord+DOWN+DOWN] = Air(self.coord+DOWN+DOWN)
            open_faces += 1
            self.open_face_coords.append(self.coord+DOWN)
        if not CUBE_ARRAY.get(self.coord+LEFT,None):
            AIR_ARRAY[self.coord+LEFT] = Air(self.coord+LEFT)
            if not CUBE_ARRAY.get(self.coord+LEFT+LEFT,None):
                AIR_ARRAY[self.coord+LEFT+LEFT] = Air(self.coord+LEFT+LEFT)
            open_faces += 1
            self.open_face_coords.append(self.coord+LEFT)
        if not CUBE_ARRAY.get(self.coord+RIGHT,None):
            AIR_ARRAY[self.coord+RIGHT] = Air(self.coord+RIGHT)
            if not CUBE_ARRAY.get(self.coord+RIGHT+RIGHT,None):
                AIR_ARRAY[self.coord+RIGHT+RIGHT] = Air(self.coord+RIGHT+RIGHT)
            open_faces += 1
            self.open_face_coords.append(self.coord+RIGHT)
        if not CUBE_ARRAY.get(self.coord+FORWARD,None):
            AIR_ARRAY[self.coord+FORWARD] = Air(self.coord+FORWARD)
            if not CUBE_ARRAY.get(self.coord+FORWARD+FORWARD,None):
                AIR_ARRAY[self.coord+FORWARD+FORWARD] = Air(self.coord+FORWARD+FORWARD)
            open_faces += 1
            self.open_face_coords.append(self.coord+FORWARD)
        if not CUBE_ARRAY.get(self.coord+BACKWARD,None):
            AIR_ARRAY[self.coord+BACKWARD] = Air(self.coord+BACKWARD)
            if not CUBE_ARRAY.get(self.coord+BACKWARD+BACKWARD,None):
                AIR_ARRAY[self.coord+BACKWARD+BACKWARD] = Air(self.coord+BACKWARD+BACKWARD)
            open_faces += 1
            self.open_face_coords.append(self.coord+BACKWARD)
        return open_faces
    
    @property
    def true_open_faces(self):
        global counted_air_coords
        faces = 0
        for coord in self.open_face_coords:
            if AIR_ARRAY[coord].group_id == AIR_ARRAY[Coord(x=22, y=12, z=12)].group_id:
                faces+=1
                counted_air_coords.append(coord)
        return faces
    
```

```python
cubes = {Cube(Coord(*line.split(","))) for line in q_input}

CUBE_ARRAY = {cube.coord:cube for cube in cubes}

part_1 = 0
for cube in cubes:
    part_1 += cube.open_faces

print("Part 1:", part_1)


for air in AIR_ARRAY.values():
    air.set_group_id()

part_2 = 0
for cube in cubes:
    part_2 += cube.true_open_faces

print("Part 2:", part_2)
```
