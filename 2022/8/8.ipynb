```python
with open("8.txt") as i:
    q_input = i.read().splitlines()
```

```python
class Tree:
    def __init__(self, coord, height):
        self.coord = coord
        self.height = int(height)
        self.vision = 0
    
    def __eq__(self, other):
        return self.coord == other.coord
    def __hash__(self):
        return hash((self.coord, self.height))
    
    def __lt__(self, other):
        return self.height < other.height

    def __le__(self, other):
        return self.height <= other.height

    def __str__(self):
        return f"{self.height}"
    def __repr__(self):
        return f"Tree(coord={self.coord}, height={self.height})"

GRID = []

for y, row in enumerate(q_input):
    GRID.append([Tree((x,y), height) for x, height in enumerate(row)])

g_width = len(GRID[0])
g_height = len(GRID)
```

```python
visible_trees = set()
visible_count = 0

# TOP DOWN
for x in range(g_width):
    current_column = []
    for y in range(g_height):
        current_tree = GRID[y][x]
        if x == 0:
            visible_trees.add(current_tree)
            continue
        if y == 0 or current_tree > max(current_column):
            visible_trees.add(current_tree)
        current_column.append(current_tree)


# BOTTOM UP
for x in reversed(range(g_width)):
    current_column = []
    for y in reversed(range(g_height)):
        current_tree = GRID[y][x]
        if x == 0:
            visible_trees.add(current_tree)
            continue
        if y == g_height-1 or current_tree > max(current_column):
            visible_trees.add(current_tree)
        current_column.append(current_tree)

#LEFT RIGHT
for y in range(g_height):
    current_row = []
    for x in range(g_width):
        current_tree = GRID[y][x]
        if y == 0:
            visible_trees.add(current_tree)
            continue
        if x == 0 or current_tree > max(current_row):
            visible_trees.add(current_tree)
        current_row.append(current_tree)

# RIGHT LEFT
for y in reversed(range(g_height)):
    current_row = []
    for x in reversed(range(g_width)):
        current_tree = GRID[y][x]
        if y == 0:
            visible_trees.add(current_tree)
            continue
        if x == g_width-1 or current_tree > max(current_row):
            visible_trees.add(current_tree)
        current_row.append(current_tree)

print(f"Part One: {len(visible_trees)}")


#VISUALISATION
#import copy

# V_GRID = copy.deepcopy(GRID)

# for y in range(g_height):
#     print()
#     for x in range(g_width):
#         tree = V_GRID[y][x]
#         if tree in visible_trees:
#             V_GRID[y][x] = "_"
        
#         print(V_GRID[y][x], end="")

```

```python
best_score = 0

for tree in visible_trees:
    uc, rc, dc, lc = True, True, True, True
    uv, rv, dv, lv = 0,0,0,0
    x, y = tree.coord
    distance = 1
    while uc or rc or dc or lc:
        try:
            if uc:
                if y-distance >= 0 and GRID[y-distance][x] < tree:
                    uv += 1
                elif y-distance >= 0 and GRID[y-distance][x] >= tree:
                    uv += 1
                    uc = False
                else:
                    uc = False
        except IndexError:
            uc = False
        try:
            if rc:
                if GRID[y][x+distance] < tree:
                    rv += 1
                elif GRID[y][x+distance] >= tree:
                    rv += 1
                    rc = False
                else:
                    rc = False
        except IndexError:
            rc = False
        try:
            if dc:
                if GRID[y+distance][x] < tree:
                    dv += 1
                elif GRID[y+distance][x] >= tree:
                    dv += 1
                    dc = False
                else:
                    dc = False
        except IndexError:
            dc = False
        try:
            if lc:
                if x-distance >= 0 and GRID[y][x-distance] < tree:
                    lv += 1
                elif x-distance >= 0 and GRID[y][x-distance] >= tree:
                    lv += 1
                    lc = False
                else: 
                    lc = False
        except IndexError:
            lc = False
        distance +=1
    
    score = uv * rv * dv * lv
    if score > best_score:
        best_score = score


print(f"Part Two: {best_score}")
    
```
