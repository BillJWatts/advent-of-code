```python
from collections import deque

with open("input.txt") as f:
    grid = [list(line) for line in f.read().splitlines()]

grid_size = len(grid)

def grid_get(coord):
    y,x = coord
    if y < 0 or x < 0:
        return None
    if y >= grid_size or x >= grid_size:
        return None
    
    return grid[y][x]

def flood_fill(coord):
    deq = deque()
    
    deq.append(coord)
    
    char = grid_get(coord)
    
    travelled_nodes = set()
    
    while deq:
        n = deq.popleft()
        
        n_char = grid_get(n)
        
        if not n_char:
            continue
        
        if n_char != char:
            continue
        
        if n in travelled_nodes:
            continue
        
        ny, nx = n
        nl = (ny, nx-1)
        nr = (ny, nx+1)
        nu = (ny-1, nx)
        nd = (ny+1, nx)
        deq.extend([nl,nr,nu,nd])
        
        travelled_nodes.add(n)
        
    return list(travelled_nodes)
        
def border_length(coords):
    length = 0
    for ny,nx in coords:
        nl = (ny, nx-1)
        nr = (ny, nx+1)
        nu = (ny-1, nx)
        nd = (ny+1, nx)
        length += 4-len([coord for coord in coords if coord in [nl, nr, nu, nd]])
    return length

def number_of_sides(coords):
    sides = 0
    for ny,nx in coords:
        nl = (ny, nx-1) in coords
        nr = (ny, nx+1) in coords
        nu = (ny-1, nx) in coords
        nd = (ny+1, nx) in coords
        nlu = (ny-1, nx-1) in coords
        nru = (ny-1, nx+1) in coords
        nld = (ny+1, nx-1) in coords
        nrd = (ny+1, nx+1) in coords
        
        if not any([nl,nr,nu,nd]):
            sides+=4
            continue
        
        if all([nl,nr,nu,nd,nlu,nru,nld,nrd]):
            continue
        
        if nl and nr and not nu and not nd:
            continue
        
        if nu and nd and not nl and not nr:
            continue
        
        if not nl and not nu:
            sides+=1
        elif nl and nu and not nlu:
            sides+=1
            
        if not nu and not nr:
            sides+=1
        elif nu and nr and not nru:
            sides+=1
        
        if not nr and not nd:
            sides+=1
        elif nr and nd and not nrd:
            sides+=1
            
        if not nd and not nl:
            sides+=1
        elif nd and nl and not nld:
            sides+=1
        
    
    return sides   

    
        
travelled_nodes = []
price = 0

for y, row in enumerate(grid):
    for x, char in enumerate(row):
        if (y,x) in travelled_nodes:
            continue
        flood_nodes = flood_fill((y,x))
        fence_sides = number_of_sides(flood_nodes)
        price += len(flood_nodes) * fence_sides
        travelled_nodes.extend(flood_nodes)
        
print("Fence price:", price)
```
