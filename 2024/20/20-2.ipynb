```python
with open("example.txt") as f:
    grid = [list(line) for line in f.read().splitlines()]
    
start = [(x, y) for y, row in enumerate(grid) for x, cell in enumerate(row) if cell == "S"][0]
end = [(x, y) for y, row in enumerate(grid) for x, cell in enumerate(row) if cell == "E"][0]

def get_neighbors(node):
    x, y = node
    for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
        nx, ny = x + dx, y + dy
        if grid[ny][nx] in [".", "E"]:
            return nx, ny

steps = 0
cx, cy = start
costs = {(cx, cy): 0}
while True:
    nx, ny = get_neighbors((cx, cy))
    if grid[ny][nx] == "E":
        steps += 1
        costs[(cx, cy)] = steps
        break
    grid[cy][cx] = "X"
    cx, cy = nx, ny
    steps += 1
    costs[(cx, cy)] = steps


def mh_dist(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

cuts = 0
for step, cost in costs.items():
    for other_step, other_cost in costs.items():

        cut_value = other_cost - cost
        if cut_value == 74 and mh_dist(step, other_step) <= 20:
            print(step, other_step)
            print(cost, other_cost)
            cuts += 1

# print(costs)
print(cuts)
```
