```python
from queue import PriorityQueue
with open("input.txt") as f:
    grid = [list(line) for line in f.read().splitlines()]

def mh_dist(node1,node2):
    x1,y1=node1
    x2,y2=node2

    return abs(x1-x2) + abs(y1-y2)

def astar(grid):
    queue = PriorityQueue()
    g_scores = {}
    f_scores = {}

    start = (1,len(grid)-2)
    end = (len(grid)-2,1)

    for y, row in enumerate(grid):
        for x, _ in enumerate(row):
            g_scores[(x,y)] = float("inf")
            f_scores[(x,y)] = float("inf")

    g_scores[start] = 0
    f_scores[start] = mh_dist(start, end)

    queue.put((mh_dist(start, end), mh_dist(start, end), start, "R"))
    paths = {}
    backtraces = []
    while not queue.empty():
        _, _, curr_cell, incoming_direction = queue.get()
        cx, cy = curr_cell
        cell = curr_cell
        backtrace = []
        while cell!=start:
            backtrace.append(cell)
            cell = paths[cell]
        backtrace.append(start)
        backtraces.append(backtrace)
        if curr_cell==end:
            return paths, backtraces

        for d in "UDLR":
            if d == "U":
                child_cell = (cx, cy-1)
            if d == "D":
                child_cell = (cx, cy+1)
            if d == "L":
                child_cell = (cx-1, cy)
            if d == "R":
                child_cell = (cx+1, cy)
            
            if grid[child_cell[1]][child_cell[0]] == "#":
                continue

            if d == incoming_direction:
                temp_g_score = g_scores[curr_cell]+1
            else:
                temp_g_score = g_scores[curr_cell]+1001
                
            temp_f_score = temp_g_score+mh_dist(child_cell, end)
            
            if temp_f_score < f_scores[child_cell] or temp_f_score == f_scores[child_cell]+1000:
                g_scores[child_cell]=temp_g_score
                f_scores[child_cell]=temp_f_score
                queue.put((temp_f_score, mh_dist(child_cell, end), child_cell, d))
                paths[child_cell] = curr_cell
    return paths, backtraces

start = (1,len(grid)-2)
end = (len(grid)-2,1)
    
path, backtraces = astar(grid)
r0 = [trace for trace in backtraces if trace[0] == end and trace[-1]==start][0]
backtraces = [trace for trace in backtraces if len(trace)<len(r0)]
seats = set()
seats.update(r0)
for i, node in enumerate(r0):
    if i == len(r0)-1:
        break
    branches = [trace for trace in backtraces if len(trace)>=2 and trace[0]==node and trace[1]!=r0[i+1]]
    for branch in branches:
        if len(branch) == len(r0[i:]):
            seats.update(branch)

for x,y in seats:
    grid[y][x] = "O"
    
with open("grid.txt", "w") as f:
    for row in grid:
        f.write("".join(row))
        f.write("\n")
        
```
