```python
with open("input.txt", "r") as file:
    data = file.read().splitlines()
    
ranges = [[int(i) for i in line.split("-")] for line in data if "-" in line]

ranges.sort(key=lambda x: x[0])

last_i = len(ranges) -1

total = 0

min_start = 0
max_end = 0

for i, (start, end) in enumerate(ranges):
    
    if not min_start and not max_end:
        min_start = start
        max_end = end
        continue
    
    if start <= max_end:
        if end > max_end:
            max_end = end   
    elif start > max_end:
        total += max_end - min_start + 1
        min_start = start
        max_end = end
    
    if i == last_i:
        total += max_end - min_start + 1
    
    
print(f"Total IDs: {total}")
```
