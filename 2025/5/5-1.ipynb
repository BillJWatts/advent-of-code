```python
with open("input.txt", "r") as file:
    data = file.read().splitlines()
    
ranges = [line for line in data if "-" in line]
ingredients = [line for line in data if line.isnumeric()]

fresh_ingredients = 0

for ingredient in ingredients:
    ingredient_value = int(ingredient)
    for r in ranges:
        lower, upper = map(int, r.split("-"))
        if lower <= ingredient_value <= upper:
            fresh_ingredients += 1
            break
print(f"Fresh ingrediants: {fresh_ingredients}")
```
