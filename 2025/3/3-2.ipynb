```python
with open("input.txt", "r") as file:
    banks = file.read().split("\n")
    
output_joltage = 0

for bank in banks:
    batteries = [int(batt) for batt in bank]
    joltage = ""    
    while True:
        if len(joltage) == 12:
            break
        split_index = -11+len(joltage)
        if split_index:
            next_value = max(batteries[:-11+len(joltage)])
        else:
            next_value = max(batteries)
        joltage += str(next_value)
        index = batteries.index(next_value)
        batteries = batteries[index+1:]
        
        if len(batteries) == 12 - len(joltage):
            joltage += "".join(str(batt) for batt in batteries)
            break
    output_joltage += int(joltage)

print(f"Output joltage: {output_joltage}")
```
