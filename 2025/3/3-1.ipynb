```python
with open("input.txt", "r") as file:
    banks = file.read().split("\n")
    
output_joltage = 0
    
for bank in banks:
    batteries = [int(batt) for batt in bank]
    largest = max(batteries[:-1])
    index = batteries.index(largest)
    next_largest = max(batteries[index+1:])
    output_joltage += int(str(largest)+str(next_largest))
    
print(f"Output joltage: {output_joltage}")
```
