```python
with open ("input.txt", "r") as file:
    data = file.read().splitlines()
    
dial_value = 50
zero_counts = 0

for rotation in data:
    starting_dial_value = dial_value
    if rotation[0] == "L":
        dial_value -= int(rotation[1:])
    elif rotation[0] == "R":
        dial_value += int(rotation[1:])
    new_zero_counts = abs(dial_value//100)
    if dial_value < 0 and starting_dial_value==0 and new_zero_counts > 0:
        new_zero_counts -= 1
    left_spin = dial_value <= 0
    dial_value %= 100  
    if left_spin and dial_value == 0:
        new_zero_counts += 1
    zero_counts += new_zero_counts
    
print(f"Number of times dial hit zero: {zero_counts}")
```
