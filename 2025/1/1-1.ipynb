```python
with open ("input.txt", "r") as file:
    data = file.read().splitlines()
    
dial_value = 50
zero_counts = 0

for rotation in data:
    if rotation[0] == "L":
        dial_value -= int(rotation[1:])
    elif rotation[0] == "R":
        dial_value += int(rotation[1:])
    dial_value %= 100
    if dial_value == 0:
        zero_counts += 1

print(f"Number of times dial hit zero: {zero_counts}")
```
