```python
with open ("input.txt", "r") as file:
    id_ranges = file.read().split(",")

id_ranges = [r.split("-") for r in id_ranges]

invalid_id_sum = 0

for start, end in id_ranges:
    for i in range(int(start), int(end) + 1):
        i = str(i)
        len_i = len(i)
        if len_i % 2 != 0:
            continue
        half_len = len_i // 2
        first_half = i[:half_len]
        second_half = i[half_len:]
        if first_half == second_half:
            invalid_id_sum += int(i)
            
print(f"Sum of invalid IDs: {invalid_id_sum}")
```
