```python
with open ("input.txt", "r") as file:
    id_ranges = file.read().split(",")

id_ranges = [r.split("-") for r in id_ranges]

invalid_id_sum = 0

def split_string(division, string):
    len_string = len(string)#
    if len_string % division != 0:
        return []
    half_len = len_string // division
    parts = []
    for i in range(division):
        parts.append(string[i*half_len:(i+1)*half_len])
    return parts

for start, end in id_ranges:
    for i in range(int(start), int(end) + 1):
        i = str(i)
        len_i = len(i)
        for division in range(2, len_i + 1):
            parts = split_string(division, i)
            if parts and all(part == parts[0] for part in parts):
                invalid_id_sum += int(i)
                break
            
print(f"Sum of invalid IDs: {invalid_id_sum}")
```
